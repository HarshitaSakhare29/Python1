```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df2 = pd.read_csv("https://raw.githubusercontent.com/Premalatha-success/Datasets/main/titanic-training-data.csv")
df2.head()
```

```python
sns.catplot(x = "Sex", y = "PassengerId", data = df2)
```

```python
sns.catplot(x = "PassengerId", y = "Fare", jitter = False, data = df2)
```

```python
sns.catplot(x = "PassengerId", y = "Fare", hue="Sex", data = df2)
```

```python
sns.catplot(x = "Sex", y = "Fare", kind="swarm", data = df2)
```

```python
sns.catplot(x="Sex", y="Pclass", hue="Age", kind = "swarm", data=df2)
```

```python
sns.catplot(x="Sex", y="Fare", hue = "PassengerId", kind = "swarm", data=df2)
```

```python
sns.catplot(x="Sex", y="Fare", kind = "box", data=df2)
```

```python
sns.catplot(x="Sex", y="Fare", hue = "Survived", kind = "box", data=df2)
```

```python

sns.catplot(x="Sex", y="Fare", hue = "Survived", kind = "violin", data=df2)
```

```python
sns.catplot(x="Sex", y="Fare", hue = "Survived", kind = "violin", split = True, data=df2)
```

```python
sns.catplot(x="Sex", y="Fare", hue = "Survived", kind = "bar", data=df2)
```

```python
sns.catplot(x="Sex", y="Fare", hue = "Survived", kind = "point", data=df2)
```

```python
sns.catplot(x="Sex", y="Fare", hue="Survived",
            col="Pclass", aspect=.9,
            kind="swarm", data=df2);
```

```python
sns.distplot(df2.Pclass)
```

```python
sns.distplot(df2.Age, kde=False, rug = True)
```

```python
sns.jointplot(x="Fare", y="Age", data=df2);
```

```python
sns.jointplot(x=df2.Fare, y=df2.Age, kind="hex", data = df2)
```

```python
sns.jointplot(x="Age", y="Fare", data=df2, kind="kde");
```

```python
corrmat = df2.corr()
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(corrmat, vmax=.8, square=True)
```
